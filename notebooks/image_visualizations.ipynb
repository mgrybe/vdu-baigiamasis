```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.image as mpimg
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
import os

def plot_triple_comparison(img_path_ref, img_path_sharp, crop_coords, zoom_scale=0.3):
    # 1. Load images
    img_ref = mpimg.imread(img_path_ref)
    img_sharp = mpimg.imread(img_path_sharp)

    y1, y2, x1, x2 = crop_coords

    # Calculate aspect ratio for the figure to prevent white gaps
    h, w = img_ref.shape[:2]
    # Total width is 2w, height is h.
    fig_aspect = h / (2 * w)

    # 2. Setup figure - Use constrained_layout=False to manually control gaps
    # Adjust figsize so the height/width ratio matches your image data
    fig, axes = plt.subplots(1, 2, figsize=(20, 20 * fig_aspect),
                             gridspec_kw={'wspace': 0, 'hspace': 0})

    images = [img_ref, img_sharp]
    labels = ["Originali autorių", "Eksperimento"]
    interp_methods = ['nearest', 'nearest']

    for i in range(2):
        ax = axes[i]
        curr_img = images[i]

        scale_y = curr_img.shape[0] / img_ref.shape[0]
        scale_x = curr_img.shape[1] / img_ref.shape[1]

        sy1, sy2 = int(y1 * scale_y), int(y2 * scale_y)
        sx1, sx2 = int(x1 * scale_x), int(x2 * scale_x)

        # Main Image Display
        # 'auto' aspect allows the image to fill the subplot area entirely
        ax.imshow(curr_img, interpolation=interp_methods[i], aspect='auto')
        ax.axis('off')

        # Label in bottom-right
        ax.text(0.98, 0.02, labels[i],
                transform=ax.transAxes,
                color='white',
                fontsize=16,
                fontweight='bold',
                ha='right', va='bottom',
                bbox=dict(facecolor='black', alpha=0.5, edgecolor='none', pad=3))

        # Red crop indicator
        rect = patches.Rectangle((sx1, sy1), sx2-sx1, sy2-sy1,
                                 linewidth=1.5, edgecolor='red', facecolor='none')
        ax.add_patch(rect)

        # 3. Inset
        ax_ins = inset_axes(ax, width=f"{zoom_scale*100}%",
                            height=f"{zoom_scale*100}%", loc=3, borderpad=0.5)

        crop = curr_img[sy1:sy2, sx1:sx2]
        z_interp = 'bicubic' if i == 1 else 'nearest'
        ax_ins.imshow(crop, interpolation=z_interp)

        ax_ins.set_xticks([])
        ax_ins.set_yticks([])
        for spine in ax_ins.spines.values():
            spine.set_edgecolor('white')
            spine.set_linewidth(2)

    # Use subplots_adjust to force the bounding box to the edges
    plt.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0, hspace=0)

    output_folder = "../deleteme"
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    save_path = os.path.join(output_folder, "comparison_result.png")

    # Use bbox_inches='tight' and pad_inches=0 to ensure no white borders are added during save
    plt.savefig(save_path, bbox_inches='tight', pad_inches=0, dpi=300)

    plt.show()
```

```python
plot_triple_comparison(
    '../demo/images/original-realsr-sharper/Canon_041.png',
    '../demo/images/improved-realsr-sharper/Canon_041.png',
    (0, 100, 500, 600), zoom_scale=0.5
)
```
